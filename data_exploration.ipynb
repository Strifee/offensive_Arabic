# Dataset 2

```python
df2 = pd.read_csv('data/ajcomments.csv',sep=",")
df2
```

```python
df2=df2.loc[:,['body','languagecomment']]
df2
```

```python
df2_off = df2.loc[df2['languagecomment'] != 0, ['body', 'languagecomment']]
df2_not_off = df2.loc[df2['languagecomment'] == 0, ['body', 'languagecomment']]
```

```python
df2_off = df2_off.drop(columns='languagecomment')
df2_off.columns = ['Comment']
df2_off
```

```python
df2_not_off = df2_not_off.drop(columns='languagecomment')
df2_not_off.columns = ['Comment']
df2_not_off
```

```python
not_offensive_df = pd.concat([df_not_off, df2_not_off])
```

```python
not_offensive_df
```

```python
offensive_df = pd.concat([df_off, df2_off])
```

```python
offensive_df
```

```python
not_offensive_df.to_csv("data/not_offensive_dataset.csv")
offensive_df.to_csv("data/offensive_dataset.csv")
```

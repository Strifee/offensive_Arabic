```python
import pandas as pd
```

# Dataset 1

```python
df = pd.read_csv('data/data_4000_selected.csv')
df = df.iloc[:, 2:4]
df
```

```python
df["Majority_Label"].replace({"Non-Offensive":0,"Offensive":1},inplace=True)
```

```python
df
```

```python
df['Majority_Label'].value_counts()
```

# Dataset 2

```python
df2 = pd.read_csv('data/ajcomments.csv',sep=",")
df2
```

```python
df2=df2.loc[:,['body','languagecomment']]
df2
```

```python
df2_off = df2.loc[df2['languagecomment'] != 0, ['body', 'languagecomment']]
df2_not_off = df2.loc[df2['languagecomment'] == 0, ['body', 'languagecomment']]
```

```python
df2_off = df2_off.drop(columns='languagecomment')
df2_off.columns = ['Comment']
df2_off
```

```python
df2_off = df2_off.assign(Majority_Label = lambda x: 1)
```

```python
df2_off
```

```python
df2_not_off = df2_not_off.drop(columns='languagecomment')
df2_not_off.columns = ['Comment']
df2_not_off
```

```python
df2_not_off = df2_not_off.assign(Majority_Label = lambda x: 0)
```

```python
df2 = pd.concat([df2_off, df2_not_off])
```

```python
df2
```

```python
df2['Majority_Label'].value_counts()
```

# Final Dataset

```python
dff = pd.concat([df, df2])
```

```python
dff
```

```python
dff['Majority_Label'].value_counts()
```

```python
dff = dff.sample(frac = 1)
dff.reset_index(drop=True, inplace=True)
```

```python
dff
```

```python
dff.to_csv("data/offensivelang_dataset.csv", index=False)
```
